## **Class Chromosome**

```
#Chromosome 
class Chromosome:

  def __init__(self):
    self.chromosome = [] #list of genes in the chromosome, and each gene has professor name,section,course,day,time,room number
    self.hconstraint = 6
    self.sconstraint = 4
    self.tconflict = dict()
    self.cconflict = dict()
    self.rconflict = dict()    
    pass


  def Representation(self):
    data = {'TimeSlots':['8:30-9:50','10:00-11:20','11:30-12:50','1:00-2:20','2:30-3:50','4:00-5:00'],
            '302':[' ',' ',' ',' ',' ',' '],
            '303':[' ',' ',' ',' ',' ',' '],
            '304':[' ',' ',' ',' ',' ',' '],
            '305':[' ',' ',' ',' ',' ',' '],
            '306':[' ',' ',' ',' ',' ',' '],
            '307':[' ',' ',' ',' ',' ',' '],
            '308':[' ',' ',' ',' ',' ',' '],
            '309':[' ',' ',' ',' ',' ',' '],
            '310':[' ',' ',' ',' ',' ',' ']
            }
    frame = pd.DataFrame(data)
    return frame


#(Professor name,Section,Course, Room Number)
  #This function initializes each chromosome by using above dataset
  def InitializeChromosome(self):
    size = df.shape[0]-1
    for index, row in df.iterrows():
      room = random.choice(Rooms)
      day = random.choice(Days)
      cell = random.choice(time)
      while(True):
        if((day == 'Friday' and cell == '1:00-2:20') or (day == 'Monday' and cell == '8:30-9:50')):
          day = random.choice(Days)
          cell = random.choice(time)
        else:
          break
      d = Individual(row.Professor,row.Section,row.Course,day,cell,room)
      self.chromosome.append(d)
    self.LAbBeforeLecture()


  def printChromosome(self):
    for i in range(len(self.chromosome)):
      self.chromosome[i].printIndividual()
    print()

  """This Function calculates Two things:
  1. No teacher can hold two classes at the same time
  2. No teacher can hold three consecutive classes """
  def teacherConflicts(self,ch): #for each chromosome in the population
    conflicts = dict()
    consecutive = dict()
    for name in (Professor):
      #this dictionary will be created to store key-value: day-times for a particular professor
      #and then each day values will be checked for duplicate times, and is there exists any duplicate
      #times then conflict has found
      filtered =dict()
      for j in range(len(ch.chromosome)):
        if(ch.chromosome[j].Pname == name):
          #first check if the day exists in the dictionary
          if ch.chromosome[j].day in filtered:
            filtered[ch.chromosome[j].day].append(ch.chromosome[j].time)
          else:
            filtered[ch.chromosome[j].day] = [ch.chromosome[j].time]
      #Now check for the existence of conflicts:
      count = 0

      """For teachers holding more than two classes simaltaneously"""
      for row in filtered.values():
        #count number of duplicates which means conflicts
        a_set = set(row)
        if(len(row) != len(a_set)):
          conflicts[name] = 'True'
          break
        else:
          conflicts[name] = 'False'
      """For teachers holding three consecutive classes""" 
      for row in filtered.values():
        if(len(row) >= 3):   #if a teacher has equals to or more than 3 classes, then constraint can be voilated
          for value in row:  #In case above condition is true, then check for each value for the possibility of 3 consec classes
            ind=time.index(value)
            if(ind+2 <= len(time)-1): #If the time slot has 2 more slots ahead it
              if(time[ind+1] in row and time[ind+2]):
                consecutive[name] = 'True'
                break
        if name not in filtered:
          consecutive[name] = 'False'

    #print("\t***Conflict of Teacher that No teacher can hold two classes at the same time***")
    #print(conflicts)
    #print("\t***Conflict of Teacher that No teacher can hold three classes consecutively***")
    #print(consecutive)
    return conflicts,consecutive


  #No section can listen for two classes at the same time
  def sectionConflicts(self,ch): #for each chromosome in the population
    conflicts = dict()
    for section in (Sections):
      #this dictionary will be created to store key-value: day-times for a particular professor
      #and then each day values will be checked for duplicate times, and is there exists any duplicate
      #times then conflict has found
      filtered =dict()
      for j in range(len(ch.chromosome)):
        if(ch.chromosome[j].Class == section):
          #first check if the day exists in the dictionary
          if ch.chromosome[j].day in filtered:
            filtered[ch.chromosome[j].day].append(ch.chromosome[j].time)
          else:
            filtered[ch.chromosome[j].day] = [ch.chromosome[j].time]
      #Now check for the number of conflicts:
      count = 0
      for row in filtered.values():
        #count number of duplicates which means conflicts
        a_set = set(row)
        if(len(row) != len(a_set)):
          conflicts[section] = 'True'
          break
        else:
          conflicts[section] = 'False'
    #print("\t***Conflict of section that No section can hold two classes at the same time***")
    #print(conflicts)
    return conflicts


  #No classroom can hold two classes at the same time
  def roomConflicts(self,ch): #for each chromosome in the population
    conflicts = dict()
    for room in (Rooms):
      #this dictionary will be created to store key-value: day-times for a particular room
      #and then each day values will be checked for duplicate times, and is there exists any duplicate
      #times then conflict has found
      filtered =dict()
      for j in range(len(ch.chromosome)):
        if(ch.chromosome[j].room == room):
          #first check if the day exists in the dictionary
          if ch.chromosome[j].day in filtered:
            filtered[ch.chromosome[j].day].append(ch.chromosome[j].time)
          else:
            filtered[ch.chromosome[j].day] = [ch.chromosome[j].time]
      #Now check for the number of conflicts:
      count = 0
      for row in filtered.values():
        #count number of duplicates which means conflicts
        a_set = set(row)
        if(len(row) != len(a_set)):
          conflicts[room] = 'True'
          break
        else:
          conflicts[room] = 'False'
    #print("\t***Conflict of location that No room can hold two classes at the same time***")
    #print(conflicts)
    return conflicts

#****************************************************************#
  def LAbBeforeLecture(self):
    #No lab should take place before the Lecture:
    j=0
    for row1 in self.chromosome:
      if(row1.subject in Labs):
        splits = row1.subject.split()
        sname = splits[0]
        i = 0
        for row2 in self.chromosome:   
          if(self.chromosome[i].subject == sname and row1.Class == self.chromosome[i].Class):
            if(Days.index(row2.day) > Days.index(row1.day)):
              self.chromosome[j].time = row2.time
              self.chromosome[j].day = row2.day
              self.chromosome[i].time = time[time.index(row2.time)-1]
              self.chromosome[i].day = Days[Days.index(row2.day)-1]
          i+=1
      j+=1

  #This function will calculate the fitness of each Chromosome
  def CalculateFitness(self,ch):
    #Number of conflicts in the "No teacher can hold two classes at the same time"
    #No teacher can have 3 consecutive classes
    self.tconflict,self.consec = self.teacherConflicts(ch)
    #Conflict determining for the classes that "No section can hold two classes at the same time"
    self.cconflict = self.sectionConflicts(ch)
    #Conflict determining for the location that "No room can hold two classes at the same time"
    self.rconflict = self.roomConflicts(ch)
    
    #Now lets calculate fitness value of  the chromosome:
    count=0
    if('True' in self.tconflict.values()):
      count+=1
    if('True' in self.consec.values()):
      count+=1
    if('True' in self.cconflict.values()):
      count+=1
    if('True' in self.rconflict.values()):
      count+=1
    fitness = self.hconstraint - count
    print("Fitness of the Chromosome is ",fitness+4)
    return fitness+4
```

## **Class Population**

```
class Population:
  def __init__(self,size):
    self.population = dict()
    self.psize = size

  def CreatePopulation(self):
    for i in range(self.psize):
      c = Chromosome()
      c.InitializeChromosome()
      self.population["C"+str(i)] = c

  def printPopulation(self):
    for name,chromo in self.population.items():
      print("\t*****",name,"*****")
      chromo.printChromosome()  
```

## **Class Genetic Algorithm**

```
import operator

class GeneticAlgorithm:
  def __init__(self):
    self.P = Population(4)
    self.h = 6
    self.s = 4
    self.selected = []
    self.fitness = dict()

  def getPopulation(self):
    self.P.CreatePopulation()


  """Selection of Chromosomes for the Next Generation"""
  def RouletteWheelSelection(self,fitness):
    max = sum(fitness.values())
    pick = random.uniform(0, max)
    current = 0
    for key, value in fitness.items():
        current += value
        if current > pick:
            return key
        
  def FindOffSpringFitness(self,child):
    fitness = dict()
    for name,chromo in child.population.items():
      fitness[name]=chromo.CalculateFitness(chromo) 
    print()
    return fitness

  def FindFitness(self):
    for name,chromo in self.P.population.items():
      self.fitness[name]=chromo.CalculateFitness(chromo)
    print()
    
    #Now lets perfom wheel selection to select chromosome for the nexr generation
    for i in range(self.P.psize):
      self.selected.append(self.RouletteWheelSelection(self.fitness))

#*********************************************************************************#
  """Perfrom Cross over on the chromosomes for producing next generation"""
  #Returns new children after cross over
  def CrossOver(self):
    children = Population(self.P.psize)
    i=0
    while(i<len(self.selected)):
      c1 = self.P.population[self.selected[i]]
      c2 = self.P.population[self.selected[i+1]]
      newc1 = Chromosome()
      newc2 = Chromosome()
      len1 = len(c1.chromosome)//2
      for j in range(len1):
        newc1.chromosome.append(c1.chromosome[j])
        newc2.chromosome.append(c2.chromosome[j])
      while(len1 < len(c1.chromosome)):
        newc1.chromosome.append(c2.chromosome[len1])
        newc2.chromosome.append(c1.chromosome[len1])
        len1+=1
      children.population["newC"+str(i)] = newc1
      children.population["newC"+str(i+1)] = newc2
      i+=2
    return children


#*********************************************************************************#
  #Perform Mutation
    """Mutation of the selected chromosome"""
  def Mutation(self,OffSpring,fitness):
    name,fitness = fitness[0]
    chromo = OffSpring.population[name]
    filtered = dict()
    if('True' in chromo.tconflict.values()):
      tname = None
      for key,value in chromo.tconflict.items():
        tname = key
        if(value == 'True'):
          for j in range(len(chromo.chromosome)):
            if(chromo.chromosome[j].Pname == key):
              #first check if the day exists in the dictionary
              if chromo.chromosome[j].day in filtered:
                filtered[chromo.chromosome[j].day].append(chromo.chromosome[j].time)
              else:
                filtered[chromo.chromosome[j].day] = [chromo.chromosome[j].time]
          for day,row in filtered.items():
          #count number of duplicates which means conflicts
            elemset = set()
            ctime = None
            for element in row:
              if element not in elemset:
                elemset.add(element)
              else:
                ctime = element
                for k in range(len(chromo.chromosome)):
                  if(chromo.chromosome[k].Pname == tname and chromo.chromosome[k].day == day and chromo.chromosome[k].time == ctime):
                    while(True):
                      ttime = random.choice(time)
                      if(time != ctime):
                        break
                    chromo.chromosome[k].time = ttime
                    break
                OffSpring.population[name] = chromo 
                return OffSpring
    elif('True' in chromo.cconflict.values()):
      sname = None
      for key,value in chromo.cconflict.items():
        sname = key
        if(value == 'True'):
          for j in range(len(chromo.chromosome)):
            if(chromo.chromosome[j].Class == key):
              #first check if the day exists in the dictionary
              if chromo.chromosome[j].day in filtered:
                filtered[chromo.chromosome[j].day].append(chromo.chromosome[j].time)
              else:
                filtered[chromo.chromosome[j].day] = [chromo.chromosome[j].time]
          for day,row in filtered.items():
          #count number of duplicates which means conflicts
            elemset = set()
            ctime = None
            for element in row:
              if element not in elemset:
                elemset.add(element)
              else:
                ctime = element
                for k in range(len(chromo.chromosome)):
                  if(chromo.chromosome[k].Class == sname and chromo.chromosome[k].day == day and chromo.chromosome[k].time == ctime):
                    while(True):
                      ttime = random.choice(time)
                      if(time != ctime):
                        break
                    chromo.chromosome[k].time = ttime
                    break
                OffSpring.population[name] = chromo 
                return OffSpring 
    elif('True' in chromo.rconflict.values()):
      rname = None
      for key,value in chromo.rconflict.items():
        rname = key
        if(value == 'True'):
          for j in range(len(chromo.chromosome)):
            if(chromo.chromosome[j].room == key):
              #first check if the day exists in the dictionary
              if chromo.chromosome[j].day in filtered:
                filtered[chromo.chromosome[j].day].append(chromo.chromosome[j].time)
              else:
                filtered[chromo.chromosome[j].day] = [chromo.chromosome[j].time]
          for day,row in filtered.items():
          #count number of duplicates which means conflicts
            elemset = set()
            rtime = None
            for element in row:
              if element not in elemset:
                elemset.add(element)
              else:
                rtime = element
                for k in range(len(chromo.chromosome)):
                  if(chromo.chromosome[k].room == rname and chromo.chromosome[k].day == day and chromo.chromosome[k].time == rtime):
                    while(True):
                      ttime = random.choice(time)
                      if(time != rtime):
                        break
                    chromo.chromosome[k].time = ttime
                    break
                OffSpring.population[name] = chromo 
                return OffSpring   
    return OffSpring            

#*********************************************************************************#
  def SelectChromsome(self,OffSpring,ofitness):
    #Find half of the best fit chromosome from parent and half from the children
    sorted_f1 = sorted(self.fitness.items(), key=operator.itemgetter(1),reverse=True)
    sorted_f2 = sorted(ofitness.items(), key=operator.itemgetter(1),reverse=True)
    #Now create new Population for the next generation:
    newPop = Population(self.P.psize)
    loop = 0
    for name,chromo in sorted_f1:
      newPop.population["C"+str(loop)] = self.P.population[name]
      loop += 1
      if(loop == 2):
        break
    for name,chromo in sorted_f2:
      newPop.population["C"+str(loop)] = OffSpring.population[name]
      loop += 1
      if(loop == 4):
        break
    #Now update the population for further generations
    self.P.population.clear()
    self.selected.clear()
    self.P.population.update(newPop.population)

  def returnBestFit(self,ch):
    d = Individual("PRAYER BREAK","","","FRIDAY","1:00-2:00","University MASJID")
    ch.chromosome.append(d)
    d = Individual("FACULTY MEETING","","","MONDAY","8:30-9:00",Rooms[0])
    ch.chromosome.append(d)
    return ch

  
  def FulFill_Fitness(self):
    sorted_f1 = sorted(self.fitness.items(), key=operator.itemgetter(1),reverse=True)
    #First check if all the hard constraints are satisfied
    if(self.h+self.s == sorted_f1[0][1]):

      return 'True',sorted_f1[0][0]
    else: 
      self.fitness.clear()
      return 'False',None
  
```
